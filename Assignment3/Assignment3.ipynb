```python
# Import necessary libraries for data analysis and visualization
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Load the dataset into a Pandas DataFrame
df = pd.read_csv('Heart.csv')
```

```python
# Display the first five rows of the dataset
df.head()
```

```python
df.replace('?', np.nan, inplace=True)
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
# Scatter Plot: Age vs Max Heart Rate 
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='MaxHR', palette='coolwarm')
plt.title("Scatter Plot: Age vs Max Heart Rate")
plt.xlabel("Age")
plt.ylabel("Max Heart Rate")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='AHD', y='Chol', data=df)
plt.title("Bar Plot: Average Cholesterol Level by Heart Disease")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Average Cholesterol Level")
plt.show()
```

```python
# Box Plot: Cholesterol level distribution based on presence of heart disease
plt.figure(figsize=(8, 6))
sns.boxplot(x='AHD', y='Chol', data=df, palette='Set2')
plt.title("Box Plot: Cholesterol Distribution by Heart Disease")
plt.xlabel("Heart Disease (0 = No, 1 = Yes)")
plt.ylabel("Cholesterol Level")
plt.show()
```

```python

# Line Chart: Trend of average cholesterol level by age (grouped by target)
avg_chol_by_age = df.groupby(['Age', 'AHD'])['Chol'].mean().unstack()
plt.figure(figsize=(10, 6))
avg_chol_by_age.plot(kind='line', marker='o')
plt.title("Line Chart: Average Cholesterol Level by Age and Heart Disease Status")
plt.xlabel("Age")
plt.ylabel("Average Cholesterol Level")
plt.grid(True)
plt.show()
```

```python
target_counts = df['AHD'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(target_counts, labels=['No Disease', 'Heart Disease'], autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title("Pie Chart: Heart Disease Distribution")
plt.show()
```
